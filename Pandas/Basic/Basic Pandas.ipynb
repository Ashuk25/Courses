```python
newdf2
```

```python
newdf2.loc[0,'A'] = 0.234513
```

```python
newdf2.head(2)
```

```python
newdf2 = newdf2.drop('A', axis=1) #Remove 'A' column from data
```

```python
newdf2.head()
```

```python
newdf2.loc[[331,35],[0,2]] #get more than 1 columns and rows data from dataframe
```

```python
df = pd.DataFrame({"name": ['Alfred', 'Batman', 'Alfred'],
                   "toy": [np.nan, 'Batmobile', 'Bullwhip'],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
```

```python
df
```

```python
df.dropna() #remove row which contains NaN or Null value.
```

```python
df
```

```python
df
```

```python
df.drop_duplicates(subset=['name']) # Here we used subset instead of axis for examine the column
```

```python
df.drop_duplicates(subset=['name'], keep='last') #keep last duplicate row. If we will set keep= False then whole the duplicate data remove
```

```python
df.shape
```

```python
df.info()
```

```python
df['toy'].value_counts(dropna=False)
```
