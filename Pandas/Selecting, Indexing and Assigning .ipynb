## Indexing in Pandas
    Indexing in Pandas just like indexing in array in python.
    However, pandas has its own accessor operation:
        - loc
        - iloc

### Index - Based Selection
    Selcting data based on its numerical position in data using "iloc"

### Note:
    Both "loc" and "iloc" are following row-first, column-second concept.
    It is oppositeof what we do in native python (which is column-first, row-second concept).

    TO retriew data using column or row index, we are using "iloc".
    
    for eg:
        reviews.iloc[:,   0]
                    row   column
                    index index
                    
    Here we can also pass the index in list.
        reviews.iloc[[0,1,2,...], 0]
    
    We can also use negative number in data selection.
        reviews.iloc[-5:]

```python
# Select the first value from the description column of reviews, assigning it to variable first_description

first_description = reviews['description'].iloc[0]

first_description
```

```python
# Select the first 10 values from the description column in reviews, assigning the result to variable first_descriptions.
first_descriptions = reviews['description'].iloc[:10]

first_descriptions
```

```python
# select rows from index 5 to 10 from the column 4
select_record = reviews.iloc[5:11,4]
select_record
```

### Label-Based Selection:
    This is the data index value, not its possition. It i denoted by "loc".
        review.loc[0,  'county']
                index   column
                label   label

### Note:
    "iloc" use python stdlib indexing schema. where the first element of the range is included and the last one excluded. So, 0:10 will select entries 0,....,9.
    "loc" meanwhile contain the lable not index. So 0:10 will select entries 0,...,10.  

```python
# select top 10 data from country column
country_column = reviews.loc[0:10, 'country'] # here we are using the lable of the row. It can be anything (A-Z, 0-9 characters)
country_column
```

```python
# Select the records with index labels 1, 2, 3, 5, and 8, assigning the result to the variable sample_reviews.
sample_reviews = reviews.loc[[1,2,3,5,8]]

sample_reviews
```

```python
# Create a variable df containing the country, province, region_1, and region_2 columns of the records with the index labels 0, 1, 10, and 100

cols = ['country', 'province', 'region_1', 'region_2']
indices = [0, 1, 10, 100]
df = reviews.loc[indices, cols]

df
```

```python
# Create a variable df containing the country and variety columns of the first 100 records.

df = reviews.loc[0:99,['country','variety']]

df
```

### Conditional Selection

    We can check the sepecific data present in column or not.
    
        reviews['country'] = "Italy" or reviews.loc[reviews['country'] == "Italy"]
        It return the all enteries if Italy present in the country column.
        
        Using above condintion, we can fetch 20k rows. The orignal has 130l rows. That means 15% wine originated from Italy.
    
    We can also check which one are better tha average wines. Wines are reviewed on a 80 to 100 point scale. So, this could mean wines that accurate at leat 90 points.
        We can use the (&) to bring the two question together.
        reviews.loc[(reviews['country'] == 'Italy') & (reviews['points'] >= 90)]
    
    Suppos, we will buy any wine that's made in italy or which related to above average. For this we can use pipe(|)
        reviews,loc[(reviews['country'] == 'Italy') | (reviews['points'] >= 90)]

### Pandas come up with built-in conditions:
    1) isin : It lets you select data whose value "isin" a list of values
              reviews.loc[reviews['country'].isin(['Italy', 'France'])]
    2) isnull : These method heighlight the values which are empty(NaN)
    3) notnull : These method heighlight the values which are not empty(NaN)
                for eg. To filter out wines lacking a price tag in Dataset
                    reviews.loc[reviews['price'].notnull()]
                    

```python
# Create a DataFrame italian_wines containing reviews of wines made in Italy

italian_wines = reviews.loc[reviews['country'] == 'Italy']
italian_wines
```

```python
# Create a DataFrame top_oceania_wines containing all reviews with at least 95 points (out of 100) for wines from Australia or New Zealand.

top_oceania_wines = reviews.loc[(reviews.country.isin(['Australia','New Zealand'])) & (reviews.points >= 95)]

top_oceania_wines
```

```python
# use isin builtin function for validating the data present in column or not
demo_isin = reviews.loc[reviews['country'].isin(['Italy', 'France'])]
demo_isin
```

```python
# show data which is null in price column using isnull function
null_data = reviews.loc[reviews['price'].isnull()]
null_data
```

```python
# show data which is not null in price column using notnull function
not_null_data = reviews.loc[reviews['price'].notnull()]
not_null_data
```

### Assigning Data
    We can assign either costant value or iterable values.
        reviews['critic'] = 'Everyone'
        reviews['index_backwords'] = range(len(reviews),0,-1)


```python
df = pd.DataFrame({'temp_c': [17.0, 25.0]},
                  index=['Portland', 'Berkeley'])
df
```

```python
# Where the value is a callable, evaluated on df:
df.assign(temp_f=lambda x: x.temp_c * 9 / 5 + 32)
```

```python
# Alternatively, the same behavior can be achieved by directly referencing an existing Series or sequence:
df.assign(temp_f=df['temp_c'] * 9 / 5 + 32)
```

```python
# You can create multiple columns within the same assign where one of the columns depends on another one defined within the same assign:
df.assign(temp_f=lambda x: x['temp_c'] * 9 / 5 + 32,
          temp_k=lambda x: (x['temp_f'] +  459.67) * 5 / 9)
```
